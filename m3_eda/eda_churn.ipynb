## Análise Bivariada

```python
# Hipótese: Clientes com contratos do tipo mensal são mais propensos ao churn - Quantidade
# Contruir uma Tabela de Contingência
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Hipótese: Clientes com contratos do tipo mensal são mais propensos ao churn - Proporção
# Contruir uma Tabela de Contingência
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total')
```

Considerando o público total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlação entre duas variáveis categóricas (qualitativas)

# Executar um teste de hipótese chamado Qui-Squared ou Qui-Quadrado de Pearson
# Num teste de hipóteses, duas hipóteses são formuladas:
# - Hipótese Nula (H0): Não há relação entre as variáveis (são independentes).
# - Hipótese Alternativa (H1): Há uma relação entre as variáveis (não são independentes).

# O teste serve para confirmar ou recusar a hipótese nula (H0), ou seja, se as variáveis são independentes ou não.
# Quando a probabilidade de erro é menor que 5% (0.05 ou p-value), rejeitamos a hipótese nula (H0) e aceitamos a hipótese alternativa (H1).
```

```python
# Gerar um Dataframe da Crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os Scores e P-Values do Qui-Quadrado
from scipy.stats import chi2_contingency
```

```python
# Tupla - Lista imutável
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar numeros com decimais sem a notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
#Criar Dataframe com os resultados do Qui-Quadrado
df_chi_scores_churn_contract = pd.DataFrame({'Qui-2': scores_churn_contract, 'p-value': pvalues_churn_contract}) 
```

```python
# Analisar Scores e P-Values
df_chi_scores_churn_contract
```

Conforme teste estatístico (Qui-Square), o P-value é <= 0.05. Desta forma, rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação. 

```python
# Avaliar a correlação entre uma variável categórica (qualitativa) e uma variável numérica (quantitativa)

# Usar o mesmo teste de Chi-Squared, transformando a variável quantitativa em qualitativa
#Hipótese: Clientes com menos de 6 meses de contrato são mais propensos ao churn

# Criar uma nova coluna no DataFrame com base em uma condição de outra coluna
# Coluna = 'TempoMenor6Meses'

import numpy as np
```

```python
#Criar nova coluna
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure < 6, 'Yes', 'No')
```

```python
df_churn.head(5)
```

```python
# Gerar um Dataframe da Crosstab (sem totais)
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
# Calcular os Scores
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Criar um Dataframe com os resultados do Qui-Quadrado
df_chi_scores_churn_tenure = pd.DataFrame({'Qui-2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
# Analisar os resultados do Qui-Quadrado
df_chi_scores_churn_tenure
```

Como o P-Value <= 0.05, rejeitamos a H0. Ou seja, as variáveis não são independentes. Com base no Qui2 resultante das duas análises, a correlação entre Churn e Tempo de Contrato < 6 meses é menos do que a correlação entre Churn e Tipo de Contrato.

```python
# Correlação entre duas variáveis numéricas (quantitativas)
# Tenure com TotalCharges
# A intuição diz que quanto mais tempo de contrato, maior o valor total pago pelo cliente

# Correlação entre duas vaiáveis numérias - Pearson
df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
# Correlação entre duas vaiáveis numérias - Spearman
df_churn.tenure.corr(df_churn.TotalCharges, method='spearman')
```

Há uma forte correlação entre Tenure e TotalCharges usando métodos estatísticos de correlação
