## Unificar Dataframes de Customers, Services e Contracts

```python
# Contar quantidade de registros no DataFrame
len(df_customers)
```

```python
len(df_services)
```

```python
len(df_contracts)
```

```python
# Renomear colunas
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
df_services.info()
```

```python
# Unificar Dataframe de Customers e Services
df_temp = df_customers.merge(df_services, on='IDCliente')
```

```python
df_temp.info()
```

```python
df_temp.head(5)
```

```python
# Unificar df_temp com Contracts, usando colunas de junção com nomes distintos
df_churn_temp = df_temp.merge(df_contracts, left_on='IDCliente', right_on='customerID')
```

```python
df_churn_temp.info()
```

```python
# Unificar os três DataFrames em um único DataFrame, mesmo com colunas diferentes
df_churn = df_customers.merge(df_services, on='IDCliente').merge(df_contracts, left_on='IDCliente', right_on='customerID')
```

```python
df_churn.info()
```

```python
# Removendo coluna de um DataFrame
df_churn.drop(columns=['customerID'], axis=1, inplace=True)
```

```python
df_churn.info()
```

## Inserção de Valores Ausentes
